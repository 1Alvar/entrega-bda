```python
videoGame = spark.read.csv('/user/hduser/video_game_Limpio.csv',sep=',')
videoGame = videoGame.toDF(
                            'Name',
                            'Platform',
                            'YearRelease',
                            'Genre',
                            'Publisher',
                            'NA_players',
                            'EU_players',
                            'JP_players',
                            'Other_players',
                            'Global_players',
                            'Critic_Score',
                            'Critic_Count',
                            'User_Score',
                            'User_Count',
                            'Developer',
                            'Rating'
                          )
```

```python
videoGame.show()
```

```python
videoGame.createOrReplaceTempView('videoGame')
```

```python
spark.sql('SELECT DISTINCT(Genre),rating FROM videoGame WHERE YearRelease = 2004 ORDER BY Genre').show()
```

```python
spark.sql('SELECT Name FROM videoGame WHERE YearRelease = 2004 AND rating = "M" ORDER BY Genre').show(500, False)
```

```python
spark.sql('SELECT Name,Genre,YearRelease FROM videoGame WHERE rating = "M" ORDER BY YearRelease,Genre,Name').show(500, False)
```
